### Dataset distribution graphs

#### What are these?

These graphs show the distribution of the different objects in the data we are working with.

3. Inferred Variable Types

```python
def display_graph(dataset, filename, lim):
    dataset = dataset.split('\n')
    
    dataset_keys, dataset_val = [], []
    others_keys, others = [], 0
    for s in dataset:
        a,b = s.split(' = ')
        if int(b) >= lim:
            dataset_keys.append(a)
            dataset_val.append(int(b))
        else:
            others_keys.append(a)
            others += int(b)
            
    dataset_keys.append('Others')
    dataset_val.append(others)
    
    print('Others :', others_keys)
    
    sns.set()
    sns.set_style("white")
    a4_dims = (7, 7)
    fig, ax = plt.subplots(figsize=a4_dims)
    plot = sns.barplot(ax=ax, y=dataset_keys,x=dataset_val, orient='h', color='lightskyblue')
    ax.set_xlabel('Dataset Count')
    fig.savefig('file_distribution_{}.png'.format(filename), dpi=300, bbox_inches='tight')
```

### Metrics Evaluation Graph

#### What are these?

These are the graphs showcasing the resulta of the evalution metrics that have shaped major design and code decisions for PaleoRec.

Evaluation Metrics help guage the performance of the model on the test dataset, helping us fine tune the model to produce better recommendations. The two well known evaluation metrics; Hit Ratio and Mean Reciprocal Rank(MRR) are used in our experiment. We have chosen Hit Ratio and MRR as they are good markers for the quality of recommendations provided by a Recommender System.

Three major graphs are displayed below
1. Evaluation Metrics for LSTM
2. Evaluation Metrics for GRU v/s LSTM
3. Evaluation Metrics for LSTM with Author as the user v/s LSTM with anonymous user (sans Author information)

#### How to reproduce?

Run the fang_metrics.py file available in the accuracy_calculation folder from
1. paleorec repo - for LSTM Evaluation Metrics
2. recommender repo - gru_code folder - for GRU Evaluation Metrics
3. recommender repo - author_code folder - for LSTM with Author as user Evaluation Metrics

On running the python application in each of the folders, a json file will be generated
1. lstm_metrics_data - for LSTM Evaluation Metrics
2. gru_metrics_data - for GRU Evaluation Metrics
3. lstm_author_metrics_data - for LSTM with Author as user Evaluation Metrics

Copy those files to the main recommender repo and then run the following cells to generate the above explained graphs.

```python
x = [3,5,7,10,12,14,16]
```

```python
with open('lstm_metrics_data', 'r') as metrics_file:
    lstm_eval_metrics = json.load(metrics_file)
with open('gru_metrics_data', 'r') as metrics_file:
    gru_eval_metrics = json.load(metrics_file)
```

```python
hr_data = lstm_eval_metrics['hit_ratio']
mrr_data = lstm_eval_metrics['mrr']
lstm_data_preproc_hr = pd.DataFrame({
        'Recommendation Set Size(3,5,7,10,12,14,16)': x, 
        'Proxy Observation(chain len = 2)': hr_data['y2'],
        'Interpretation Var(chain len = 3)': hr_data['y3'],
        'Proxy Units(chain len = 3)': hr_data['y3_u'],
        'Interpretation Var Detail(chain len = 4)': hr_data['y4'],
        'Inferred Var(chain len = 5)': hr_data['y5'],
        'Inferred Var Units(chain len = 6)': hr_data['y6']})
lstm_data_preproc_mrr = pd.DataFrame({
        'Recommendation Set Size(3,5,7,10,12,14,16)': x, 
        'Proxy Observation(chain len = 2)': mrr_data['y2'],
        'Interpretation Var(chain len = 3)': mrr_data['y3'],
        'Proxy Units(chain len = 3)': mrr_data['y3_u'],
        'Interpretation Var Detail(chain len = 4)': mrr_data['y4'],
        'Inferred Var(chain len = 5)': mrr_data['y5'],
        'Inferred Var Units(chain len = 6)': mrr_data['y6']})
```

#### Evaluation Metrics for LSTM

```python
dataset_list = [lstm_data_preproc_hr, lstm_data_preproc_mrr]
y_label_list = ['Hit Ratio', 'Mean Reciprocal Rank']
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,6))
# add subplots
axes_lis = [ax1, ax2]
fig_name = ['a. ', 'b. ']
for i, ds in enumerate(dataset_list):
    sns.lineplot(ax=axes_lis[i], x='Recommendation Set Size(3,5,7,10,12,14,16)', y='value', hue='variable', style="variable",
             data=pd.melt(ds, ['Recommendation Set Size(3,5,7,10,12,14,16)']), 
                markers=True, dashes=False, markersize=10)
    sns.set_style("white")
    yl = 'LSTM '
    axes_lis[i].set(ylabel=yl + '{}'.format(y_label_list[0] if i%2 == 0 else y_label_list[1]))
    axes_lis[i].set(xlabel= '{} Recommendation Set Size(3,5,7,10,12,14,16)'.format(fig_name[i]))
    axes_lis[i].get_legend().remove()
    axes_lis[i].set_ylim(0, 1.1)
    
# add legend
handles, labels = axes_lis[1].get_legend_handles_labels()
lgd = fig.legend(handles, labels, loc='lower center', ncol=3,bbox_to_anchor=(0.5,-0.1))

# add subtitle
fig.suptitle('Evaluation Metrics for LSTM')

timestr = time.strftime("%Y%m%d_%H%M%S")
fig.savefig('lstm_'+timestr+'.png', dpi=300, bbox_extra_artists=(lgd,), bbox_inches='tight')
```
